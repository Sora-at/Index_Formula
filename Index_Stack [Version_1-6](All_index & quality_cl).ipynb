```python
import pandas as pd
import numpy as np
import glob
import os
import re
import matplotlib as mpl

#import location all file
path = r'D:\Drive2\Work_Sentinal\Excel_result\Rubber\V2\Point_Extract_Pixel\Nakonpanom2022_QVE\Extract' 

filenames = glob.glob(path + "/*.xlsx")

#file count
file_count = len(filenames)
```

```python
#concat file
concat_all_sheets_all_files = pd.concat(map(pd.read_excel,filenames),axis=0)
#concat_all_sheets_all_files
```

```python
df2 = concat_all_sheets_all_files.groupby('FID') #Define column primary key for group
new_df = pd.DataFrame()
t = "_"
for i in range(file_count):#Define number total file for display name column & number
    new_df = pd.concat([new_df,df2.nth(i).add_suffix(t).add_suffix(i+1)],axis=1)    
    #n = 10 #Define number total file for break
    #if i == n-1: 
        #break

new_df
for col in new_df.columns:
    print(col)
```

```python
#input name all index
index = ["fapar", "fcover", "lai","lai_cab","lai_cw","ARVI","BI","BI2","CI","DVI","EVI","GEMI","GNDVI","IPVI",
         "IRECI","MCARI","MNDWI","MSAVI","MSAVI2","MTCI","NDI45","NDPI","NDRE","NDRE2","NDTI","NDVI","NDWI",
         "NDWI2","PSSRA","PVI","REIP","RI","RVI","S2REP","SAVI","TNDVI","TSAVI","VARI","WDVI"]
for x in index:
    df3=new_df[['No_1','Group_name_1','Code_1','E_1','N_1','Lat_1','Long_1']]#Define name column
    df4=new_df.filter(regex=  "^"+ x + "_\d" ,axis=1) #(regex=r'^BI2_\d',axis=1)#Define name index
    main_list = [s.replace('_1', '')for s in df3]
    df3.columns = main_list
#---------------------------
#merge column
    df_stack=pd.merge(df3,df4,how='left',left_on=None,right_on=None,left_index=True, right_index=True)
#---------------------------
#read name all file
    Names=next(os.walk(path))[2]
#---------------------------
#function edit string in list names of all file
    def test_as_str(Names):
        output_list = []
        for string in Names:
            output_list += re.findall(r'\d+', string)
        return output_list

    names_list=test_as_str(Names)
    
#---------------------------
#define name cloud
    df_cloud=new_df.filter(regex=r'^quality_cl_\d',axis=1)
#rename cloud column
    df_cloud.columns = names_list 
#---------------------------
#rename column
    df_stack.columns = main_list + names_list    
#---------------------------
#function background color
    def background_color(value):
        if value <= 40:
            color = 'white'
        elif value > 50:
            color = '#FF6A6A'  
        else:
            color = '#FFF68F'  
        return 'background-color: %s' % color
#---------------------------
#apply fuction background color to table
    df_final=df_stack.style.apply(lambda x: df_cloud.applymap(background_color), axis=None,subset=names_list)
#---------------------------
#export all index(.xlsx)
    df_final.to_excel('D:/Drive2/Work_Sentinal/Excel_result/Rubber/V2/Point_Extract_Pixel/Nakonpanom2022_QVE/Index/NM_QVE_2022_' + x +'.xlsx')
```
